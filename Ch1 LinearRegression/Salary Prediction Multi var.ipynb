In Excercise find the salary for the below criteria. 
1. Experience: 2 years, Test Score: 9 Interview Score: 6
2. Experience: 12 years, Test Score: 10 Interview Score: 10

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('multi_var_hiring.csv')
df
```

```python
# w2n.word_to_num(df.experience[3])

def convert_words_to_numbers(word):
    from word2number import w2n    
    try:
        return w2n.word_to_num(word)
    except ValueError:
        return word
```

```python
df['exp_nu'] = df['experience'].apply(convert_words_to_numbers)
df
```

```python
import math 
median_exp = math.floor(df.exp_nu.median())
df.exp_nu = df.exp_nu.fillna(median_exp)
df['test_score(out of 10)'] = df['test_score(out of 10)'].fillna(0)
df
```

```python
df = df.drop(columns='experience')
df
```

```python
# Rearranging columns
new_order = ['exp_nu'] + [col for col in df.columns if col != 'exp_nu']
df = df.reindex(columns=new_order)
df
```

```python
df = df.rename(columns={'exp_nu': 'experience',
                        'test_score(out of 10)':'test_marks',
                       'interview_score(out of 10)':'intervw_marks',
                       'salary($)':'salary'})
```

```python
# df.columns
df 
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['experience', 'test_marks', 'intervw_marks']],
        df.salary)
```

```python
reg.coef_
```

```python
reg.intercept_
```
