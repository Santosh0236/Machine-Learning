
1) Predict price of a mercedez benz that is 4 yr old with mileage 45000

2) Predict price of a BMW X5 that is 7 yr old with mileage 86000

3) Tell me the score (accuracy) of your model. (Hint: use LinearRegression().score())

```python
import pandas as pd 
```

```python
df = pd.read_csv("carprices.csv")
df
```

```python
dummies = pd.get_dummies(df['Car Model'])
dummies
```

```python
dummies = dummies.astype(int)
dummies
```

```python
merged = pd.concat([df,dummies],axis=1)
merged
```

```python
final = merged.drop(['Car Model'], axis=1)
final
```

```python
final = final.rename(columns={'Sell Price($)': 'Sell_Price_in_Dollar',
                              'Age(yrs)':'Age_in_Yrs',
                              'Audi A5':'Audi_A5',
                              'BMW X5':'BMW_X5',
                              'Mercedez Benz C class':'MercedezBenzCclass'})
final
```

```python
final = final.drop('MercedezBenzCclass', axis=1)
final
```

```python
X = final.drop('Sell_Price_in_Dollar', axis=1)
X
```

```python
y = final.Sell_Price_in_Dollar
y
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X,y)
```

```python
model.predict(X) 
```
