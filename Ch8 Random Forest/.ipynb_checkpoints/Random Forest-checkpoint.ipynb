```python
import pandas as pd
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
dir(digits)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.gray()
for i in range(4):
    plt.matshow(digits.images[i]);    
```

```python
df = pd.DataFrame(digits.data)
df.head()
```

```python
df['target']=digits.target
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train, y_test = train_test_split(df.drop(['target'],axis='columns'),digits.target,test_size=0.2)
```

```python
len(X_train),len(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()#n_estimator =10 Default
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
model_20tree = RandomForestClassifier(n_estimators=20)
model_20tree.fit(X_train,y_train)
model_20tree.score(X_test,y_test)
```

```python
model_40tree = RandomForestClassifier(n_estimators=40)
model_40tree.fit(X_train,y_train)
model_40tree.score(X_test,y_test)
```

```python
model_30tree = RandomForestClassifier(n_estimators=30)
model_30tree.fit(X_train,y_train)
model_30tree.score(X_test,y_test)
```

```python
y_pred_10tree = model.predict(X_test)
```

```python
y_pred_10tree
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pred_10tree)
cm
```

```python
import seaborn as sns
plt.figure(figsize=(10,7));
sns.heatmap(cm, annot=True);
plt.xlabel('Predicted');
plt.ylabel('Truth');
```

```python
y_pred_20tree = model_20tree.predict(X_test)
```

```python
cm20tree = confusion_matrix(y_test,y_pred_20tree)
cm20tree
```

```python
plt.figure(figsize=(10,7));
sns.heatmap(cm20tree, annot=True);
plt.xlabel('Predicted');
plt.ylabel('Truth');
```
