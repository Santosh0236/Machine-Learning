```python
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB
```

```python
from sklearn.datasets import load_digits
from sklearn.metrics import accuracy_score
import pandas as pd
```

```python
digit = load_digits()
dir(digit)
```

```python
df = pd.DataFrame(digit.data, columns=digit.feature_names)
df['target_name'] = digit.target
df
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(digit.data,digit.target,test_size=0.3)
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.model_selection import GridSearchCV
clf = GridSearchCV(svm.SVC(gamma='auto'),
                   { 'C':[1,10,20],
                   'kernel':['rbf','linear']
                   } , cv=5,return_train_score=False)
clf.fit(digit.data,digit.target)
clf.cv_results_
```

```python
df_cv_res =pd.DataFrame(clf.cv_results_)
df_cv_res    
```

```python
df_cv_res[['param_C','param_kernel','mean_test_score']]
```

```python
# Define the parameter grid for SVM, Random Forest, LogisticRegression,  DecisionTreeClassifier, GaussianNB, and MultinomialNB
classifiers = {
    'RandomForest': RandomForestClassifier(),
    'LogisticRegression': LogisticRegression(),
    'DecisionTree': DecisionTreeClassifier(),    
    'GaussianNB': GaussianNB(),
    'MultinomialNB': MultinomialNB()
}
param_grids = {
    'RandomForest': {'n_estimators': [10, 50, 100], 'max_depth': [None, 10, 20]},
    'LogisticRegression': {'C': [0.001, 0.01, 0.1, 1, 10, 100],'max_iter':[4000,8000,9000]},
    'DecisionTree': {'max_depth': [None, 4000, 5000, 6000]},
    'GaussianNB': {'var_smoothing': [1e-9, 1e-8, 1e-7, 1e-6]},
    'MultinomialNB':{'alpha': [0.1, 0.5, 1.0, 2.0]}
}
for name, classifier in classifiers.items():
    param_grid = param_grids[name]
    grid_search = GridSearchCV(classifier, param_grid, cv=5, scoring='accuracy')
    grid_search.fit(X_train, y_train)
    print(f"Best hyperparameters for {name}: {grid_search.best_params_}")

    # Predict on the test set using the best model
    y_pred = grid_search.predict(X_test)
    
    accuracy = accuracy_score(y_test, y_pred)
    print(f"Accuracy for {name}: {accuracy}")
```
